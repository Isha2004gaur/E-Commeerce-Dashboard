```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Monthly Revenue Trend


```python
master_df = master_df[master_df['order_purchase_timestamp'].dt.year != 2016]
```

```python
# Extract year and month from date

master_df['year_month'] = master_df['order_purchase_timestamp'].dt.to_period('M')
monthly_revenue = master_df.groupby('year_month')['payment_value'].sum().reset_index()
monthly_revenue.columns = ['month', 'revenue']

# Plot
plt.figure(figsize=(14, 5))
plt.plot(monthly_revenue['month'].astype(str),
         monthly_revenue['revenue'],
         marker='o',
         color='steelblue',
         linewidth=2)

plt.title('Monthly Revenue Trend', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Revenue (BRL)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()


monthly_revenue_sorted = monthly_revenue.sort_values('revenue', ascending=False)
print(f"Total Revenue  : {master_df['payment_value'].sum():,.2f} BRL")


print(f"Best Month  : {monthly_revenue_sorted.iloc[0]['month']} → {monthly_revenue_sorted.iloc[0]['revenue']:,.2f} BRL")
print(f"Worst Month : {monthly_revenue_sorted.iloc[-1]['month']} → {monthly_revenue_sorted.iloc[-1]['revenue']:,.2f} BRL")
```

Overall revenue shows a strong upward growth

```python
# Drop any existing 'review_score' columns to prevent MergeError
cols_to_drop = [col for col in master_df.columns if 'review_score' in col]
if cols_to_drop:
    master_df = master_df.drop(columns=cols_to_drop)

master_df = master_df.merge(reviews[['order_id', 'review_score']], on='order_id', how='left')

## Calculate delivery days first
master_df['delivery_days'] = (
    pd.to_datetime(master_df['order_delivered_customer_date']) -
    pd.to_datetime(master_df['order_purchase_timestamp'])
).dt.days

# Remove null delivery days and outliers
scatter_df = master_df.dropna(subset=['delivery_days'])
scatter_df = scatter_df[scatter_df['delivery_days'] <= 60]  # remove extreme outliers
scatter_df = scatter_df[scatter_df['payment_value'] <= 2000]  # remove extreme outliers

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(scatter_df['delivery_days'],
            scatter_df['payment_value'],
            alpha=0.3,          # transparency so overlapping dots are visible
            color='steelblue',
            s=10)               # size of each dot

plt.title('Delivery Days vs Order Value', fontsize=14)
plt.xlabel('Delivery Days')
plt.ylabel('Order Value (BRL)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\n📦 Average Delivery Days : {scatter_df['delivery_days'].mean():.1f} days")
print(f"💰 Average Order Value   : {scatter_df['payment_value'].mean():,.2f} BRL")
```

```python
# Calculate delivery days

# Drop any existing 'review_score' columns to prevent MergeError
cols_to_drop = [col for col in master_df.columns if 'review_score' in col]
if cols_to_drop:
    master_df = master_df.drop(columns=cols_to_drop)

master_df = master_df.merge(reviews[['order_id', 'review_score']], on='order_id', how='left')

master_df['delivery_days'] = (
    pd.to_datetime(master_df['order_delivered_customer_date']) -
    pd.to_datetime(master_df['order_purchase_timestamp'])
).dt.days

# Clean data
scatter_df = master_df.dropna(subset=['delivery_days', 'review_score'])
scatter_df = scatter_df[scatter_df['delivery_days'] <= 60]
scatter_df = scatter_df[scatter_df['payment_value'] <= 2000]

# Plot with color by review score
plt.figure(figsize=(11, 6))
colors = {1:'red', 2:'orange', 3:'yellow', 4:'lightgreen', 5:'steelblue'}

for score in [1, 2, 3, 4, 5]:
    subset = scatter_df[scatter_df['review_score'] == score]
    plt.scatter(subset['delivery_days'],
                subset['payment_value'],
                alpha=0.3,
                s=8,
                color=colors[score],
                label=f'Review {score}⭐')

plt.title('Delivery Days vs Order Value (by Review Score)', fontsize=14)
plt.xlabel('Delivery Days')
plt.ylabel('Order Value (BRL)')
plt.legend(title='Review Score', bbox_to_anchor=(1.05, 1))
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Instead of plotting 115k dots
# Plot AVERAGE order value per delivery day

avg_df = scatter_df.groupby('delivery_days')['payment_value'].mean().reset_index()
avg_df.columns = ['delivery_days', 'avg_order_value']

plt.figure(figsize=(11, 6))
plt.scatter(avg_df['delivery_days'],
            avg_df['avg_order_value'],
            color='steelblue',
            alpha=0.7,
            s=50)

# Add a trend line
import numpy as np
z = np.polyfit(avg_df['delivery_days'], avg_df['avg_order_value'], 1)
p = np.poly1d(z)
plt.plot(avg_df['delivery_days'],
         p(avg_df['delivery_days']),
         color='red',
         linewidth=2,
         label='Trend Line')

plt.axvline(x=10, color='green', linestyle='--', linewidth=1.5, label='Sweet Spot Start')
plt.axvline(x=40, color='red', linestyle='--', linewidth=1.5, label='Sweet Spot End')

plt.title('Avg Order Value vs Delivery Days', fontsize=14)
plt.xlabel('Delivery Days')
plt.ylabel('Avg Order Value (BRL)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

correlation = avg_df['delivery_days'].corr(avg_df['avg_order_value'])
print(f"Correlation Score : {correlation:.2f}")


```

```python
import datetime

# Reference date = 1 day after last order
reference_date = master_df['order_purchase_timestamp'].max() + datetime.timedelta(days=1)

print(f"Reference Date : {reference_date}")

# Calculate R, F, M for each customer
rfm = master_df.groupby('customer_id').agg({
    'order_purchase_timestamp' : lambda x: (reference_date - x.max()).days,  # Recency
    'order_id'                 : 'count',                                     # Frequency
    'payment_value'            : 'sum'                                        # Monetary
}).reset_index()

# Rename columns
rfm.columns = ['customer_id', 'Recency', 'Frequency', 'Monetary']

print(f"\nRFM Table Shape : {rfm.shape}")
print(f"\nFirst 5 Rows:")
rfm.head()
```

```python
# Give scores 1-5 to each customer
# Recency   → LOWER days = HIGHER score (recent buyer = good!)
# Frequency → HIGHER count = HIGHER score
# Monetary  → HIGHER spend = HIGHER score

rfm['R_Score'] = pd.qcut(rfm['Recency'],
                          q=5,
                          labels=[5, 4, 3, 2, 1])  # reversed! lower days = higher score

rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'),
                          q=5,
                          labels=[1, 2, 3, 4, 5])

rfm['M_Score'] = pd.qcut(rfm['Monetary'],
                          q=5,
                          labels=[1, 2, 3, 4, 5])

# Combine all 3 scores into one RFM score
rfm['RFM_Score'] = (rfm['R_Score'].astype(int) +
                    rfm['F_Score'].astype(int) +
                    rfm['M_Score'].astype(int))

print("Scores added! ✅")
print(f"\nRFM Score Range : {rfm['RFM_Score'].min()} to {rfm['RFM_Score'].max()}")
rfm.head()
```

```python
print(f"Minimum RFM Score : {rfm['RFM_Score'].min()}")
print(f"Maximum RFM Score : {rfm['RFM_Score'].max()}")
print(f"Average RFM Score : {rfm['RFM_Score'].mean():.2f}")
```

```python
def assign_segment(score):
    if score >= 13:
        return 'Elite'        # ← changed here!
    elif score >= 11:
        return 'Loyal'
    elif score >= 9:
        return 'Potential'
    elif score >= 6:
        return 'At Risk'
    else:
        return 'Lost'

rfm['Segment'] = rfm['RFM_Score'].apply(assign_segment)

print("Segments assigned! \n")
print("CUSTOMER SEGMENTS BREAKDOWN\n")
print(rfm['Segment'].value_counts())
```

```python
# Data
segments = rfm['Segment'].value_counts().reset_index()
segments.columns = ['Segment', 'Count']

# Colors - Custom list of hex codes
colors = [
    '#9bbfe0', # Light Blue
    '#f1d3f5', # Orange
    '#fbe29f', # Green
    '#c6d68f', # Red
    '#ffab80'  # Grey
]

# Donut Chart
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    segments['Count'],
    labels=segments['Segment'],
    colors=colors,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops=dict(width=0.5)  # ← makes it donut!
)

# Make percentage text bigger
for autotext in autotexts:
    autotext.set_fontsize(11)

# Make label text bigger
for text in texts:
    text.set_fontsize(14)

ax.set_title('RFM Customer Segmentation',
             fontsize=16,
             fontweight='bold',
             pad=20)

plt.tight_layout()
plt.show()
```
